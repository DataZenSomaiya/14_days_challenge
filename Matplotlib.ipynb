# Other 2D graphs

```python
import numpy as np
from matplotlib import pylab
n = np.array([0,1,2,3,4,5])
fig, axes = plt.subplots(1, 4, figsize=(16,5))
axes[0].set_title("scatter")
axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))

axes[1].set_title("step")
axes[1].step(n, n**2,lw=2)

axes[2].set_title("bar")
axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)

axes[3].set_title("fill_between")
axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5)
```

```python
# Radar Chart
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0, 0.0, .6, .6], polar=True)
t = np.linspace(0, 2 * np.pi, 100)
ax.plot(t,t, color='blue' ,lw=3)
```

```python
# Histogram

n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))
axes[0].set_title("Default Histogram")
axes[0].hist(n)
```

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-3.0, 3.0, delta)

x, y = np.meshgrid(x, y)
z1 = np.exp(-x**2 - y**2)
z2 = np.exp(-(x-1)**2 - (y-1)**2)
z=(z1 - z2) * 2

print(x)
print(y)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(x, y, z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title("contour map")
```

# Write a python program to create pie chart of popularity of programming languages


```python
import matplotlib.pyplot as plt
# Data to plot

language = 'Python','Java','PHP','JavaScript','C#','C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
color = ["blue", "green", "red", "violet", "yellow", "pink"]

```

```python
# explode list slice
explode = (0.1, 0, 0, 0, 0, 0)
plt.pie(popularity, explode=explode, labels=language, colors=color, autopct='%1.1f%%',shadow=True,startangle=140)
```
