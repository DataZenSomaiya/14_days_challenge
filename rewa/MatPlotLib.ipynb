```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]


plt.plot(x,y, label = '2x', color='red', linewidth=2,marker='.', markersize=12, markeredgecolor='blue')

plt.title('Our First Graph!', fontdict = {'fontname' : 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()

plt.show()
```
