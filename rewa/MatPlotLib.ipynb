```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]


plt.plot(x,y, label = '2x', color='red', linewidth=2,marker='.', markersize=12, markeredgecolor='blue')

plt.title('Our First Graph!', fontdict = {'fontname' : 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()

plt.show()
```

```python
subjects = ['Computers', 'Maths', 'Science']
marks_1 = [50,70,60]
marks_2 = [99,74,65]

plt.plot(subjects,marks_1, label = 'Xs Marks', marker ='.')
plt.plot(subjects,marks_2, label = 'Ys Marks', marker = 'x' )
plt.legend()
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks Comparison')

plt.show()
```
