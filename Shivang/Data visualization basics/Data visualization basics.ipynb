```python
import numpy as np
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv('diamonds.csv')
print(df.shape)
df.head(5)
```

```python
df.hist(column="carat",       
              figsize=(8,8),        
              color="blue");        
```

```python
df.hist(column="carat",        
              figsize=(8,8),        
              color="blue",          
              bins=50,              
              range= (0,3.5));      
```

```python
df[df["carat"] > 3.5]
```

```python
df.boxplot(column="carat");
```

```python
df.boxplot(column="price",        
                 by= "clarity",         
                 figsize= (8,8));       
```

```python
df.boxplot(column="carat",        
                 by= "clarity",         
                 figsize= (8,8));
```

```python
df["carat"].plot(kind="density",  
                      figsize=(8,8),    
                      xlim= (0,5));    
```

```python
carat_table = pd.crosstab(index=df["clarity"], columns="count")
carat_table
```

```python
carat_table.plot(kind="bar",
                 figsize=(8,8));
```

```python
carat_table = pd.crosstab(index=df["clarity"], 
                          columns=df["color"])

carat_table
```

```python
carat_table.plot(kind="bar", 
                 figsize=(8,8),
                 stacked=True);
```

```python
carat_table.plot(kind="bar", 
                 figsize=(8,8),
                 stacked=False);
```

```python
df.plot(kind="scatter",     
              x="carat",          
              y="price",          
              figsize=(10,10),
              ylim=(0,20000));
```

```python
years = [y for y in range(1950,2016)]

readings = [(y+np.random.uniform(0,20)-1900) for y in years]

time_df = pd.DataFrame({"year":years,
                        "readings":readings})

# Plot the data
time_df.plot(x="year",
             y="readings",
             figsize=(9,9));
```
