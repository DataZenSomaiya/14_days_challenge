```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

Data Import

```python
df = pd.read_csv("ds_salaries.csv")
```

```python
df.head()
```

Deleting Unwanted Column

```python
del df['Unnamed: 0']
```

Scatter Plot

```python
sns.scatterplot(data=df)
```

Histogram

```python
sns.histplot(df["salary_in_usd"], kde=True, bins=15)
```

Bar Plot

```python
sns.barplot(x="experience_level", y="salary_in_usd", data=df,palette="YlGnBu" )
```

Box Plot

```python
sns.boxplot(df, palette="YlGnBu")
```

Strip Plot

```python
sns.stripplot(x="experience_level", y="salary_in_usd", data=df, dodge=True, palette="YlGnBu" )
```

Joint Plot

```python
sns.jointplot(x="experience_level", y="salary_in_usd", data=df  )
```

Joint Plot with Regression

```python
sns.jointplot(x="remote_ratio", y="salary_in_usd", data=df ,kind="reg" )
```

Blurred Joint Plot

```python
sns.jointplot(x="remote_ratio", y="salary_in_usd", data=df ,kind="kde", shade=True, cmap="YlGnBu" )
```

PairPlot

```python
sns.pairplot(df.select_dtypes(['number']))
```

Heatmap

```python
sns.heatmap(df.corr(), annot=True)
```

Clearing object type for Clustermap

```python
test = df
```

```python
del test["experience_level"]
del test["employment_type"]
del test["job_title"]
del test["salary_currency"]
del test["employee_residence"]
del test["company_size"]
del test["company_location"]
```

Cluster Map

```python
sns.clustermap(test)
```
