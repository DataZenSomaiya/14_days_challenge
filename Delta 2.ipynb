```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [2,4,5,6,9]

plt.plot(x,y, label = '2x', color = 'red', linewidth='2', marker='.', markersize = 10, markeredgecolor = 'blue')
x2 = np.arange(0,2.5,0.6)
plt.plot(x2, x2**3,label = 'x**3', linewidth=2, marker='.', markersize = 10, markeredgecolor = 'blue')
print(x2)
plt.title(' Our First Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()
plt.savefig('my graph.png')
plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
print(gas)
```

```python
plt.title('Gas prices Over Time (in USD)')
plt.plot(gas.Year, gas.UK, label = 'UK', color = 'black', marker = '.')
plt.plot(gas.Year, gas.Canada, label = 'Canada', color = 'orange', marker = '.')
plt.plot(gas.Year, gas['South Korea'], color = 'yellow', marker = '.', label = 'South Korea')
plt.plot(gas.Year, gas.USA, label = 'USA', color = 'brown', marker = '.')
plt.legend()
#countries_to_look_at = ['USA', 'UK','Japan']
plt.xlabel=('Years')
plt.ylabel= ('Prices in UDS')
plt.legend()
print(gas.Year[::3])
plt.xticks(gas.Year[::3])
plt.show()
plt.savefig('gasprice.png')
```
